```python
# Import Required Libraries
import yfinance as yf
import pandas as pd
import pprint
```

Using the `Ticker` module to create an object that will allow us to access functions to extract data. To do this we need to provide the ticker symbol for the stock, for example: `TSLA` (Tesla)

```python
ticker_name = input("Please Enter the Sotck Ticker Symbol: ")
ticker = yf.Ticker(ticker_name)
```

Using the attribute  <code>info</code> we can extract information about the stock as a Python dictionary.

```python
ticker_info = ticker.info
# ticker_info

company_info = dict()
company_info["Ticker Symbol"] = ticker_name
company_info["Sector"] = ticker_info["sector"]
company_info["Industry"] = ticker_info["industry"]
company_info["Country"] = ticker_info["country"]
company_info["State"] = ticker_info["state"]
company_info["City"] = ticker_info["city"]
company_info["Address"] = ticker_info["address1"]
company_info["Website"] = ticker_info["website"]

pprint.pprint(company_info)
```

Using the <code>history()</code> method to get the share price of the stock over a certain period of time. Using the `period` parameter we can set how far back from the present to get data. The options for `period` are 1 day (1d), 5d, 1 month (1mo) , 3mo, 6mo, 1 year (1y), 2y, 5y, 10y, ytd, and max.

```python
ticker_price_data = ticker.history(period="1mo")
ticker_price_data

# The format that the data is returned in is a Pandas DataFrame. 
# With the Date as the index the share Open, High, Low, Close, Volume, and Stock Splits are given for each day
```

```python
# Print the data for the last 5 days
ticker_price_data.tail(5)
```

Reset the index of the DataFrame with the `reset_index` function. Also, set the `inplace` paramter to `True` so the change takes place to the DataFrame itself.


```python
ticker_price_data.reset_index(inplace=True)
ticker_price_data
```

Plot the `Open` price against the `Date`:

```python
ticker_price_data.plot(x="Date", y="Open")
```

Plot the `Close` price against the `Date`:

```python
ticker_price_data.plot(x="Date", y="Close")
```

Plot the trading `Volume` against the `Date`:

```python
ticker_price_data.plot(x="Date", y="Volume")
```
